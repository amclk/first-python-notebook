```python
merged["clean_firstname"] = merged.contributor_firstname.fillna("")
```

```python
merged.head()
```

```python
merged['clean_fullname'] = merged.clean_firstname +  " "  + merged.contributor_lastname
```

```python
bigdogs = merged.groupby(
    [
        "clean_fullname",
         "committee_position"]
).amount.sum().reset_index().sort_values(
    "amount", ascending = False
)
```

```python
bigdogs[bigdogs.amount >= 100000].amount.sum() / merged.amount.sum()
```

```python
bigdogs.head(10)
```

```python
def standardize_name(clean_fullname):
    clean_fullname = clean_fullname.strip()
    if clean_fullname == "SEAN PARKER AND AFFILIATED ENTITIES":
        return "SEAN PARKER"
    elif clean_fullname == "FUND FOR POLICY REFORM (NONPROFIT 501 (C)(4))":
        return "FUND FOR POLICY REFORM"
    else: 
        return clean_fullname
        
```

```python
bigdogs["standardized_name"] = bigdogs.clean_fullname.apply(standardize_name)
```

```python
bigdogs.head(10)
```

```python
bigdogs.head(10)
```

```python
bigdogs.head(10)
```

```python
bigdogs.groupby(
    ["standardized_name", "committee_position"]
).amount.sum().reset_index().sort_values(
    "amount", ascending = False)
```
